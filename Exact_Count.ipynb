Generating Stream of Random Numbers

```python
#@title Randomly Generating Data Stream

with open("numbers", 'w') as f:
  no_of_element = rnd.randint(900000000, 1000000000)
  for i in range(no_of_element):
    r = rnd.randint(1000, 10000)
    f.write("{0}\n".format(rnd.randint(1, r)))
```

#Exact Count

```python
file = open("/content/drive/MyDrive/My Documents/ABD/Assignment2/numbers",'r')
```

```python
count = 0
line = file.readline()  

#Keys: Unique data, value = count
start_time = time.time()
unique_data_count={}

while line:
  count = count+1
  line = line.rstrip() 

  if (unique_data_count.get(line) == None):
    unique_data_count[line] = 1
  else:
    unique_data_count[line] = int(unique_data_count.get(line))+1

  line = file.readline()
  
file.close()
end_time = time.time()
```

<b>Exact Summary

Total number of integers in the file.

```python
print("Total number of integers in the file: ",count)
print("Time(in seconds) required for Exact Count:", (end_time - start_time))
```

Total number of unique integers in the file

```python
print("No of unique items in the list are:", len(unique_data_count))
```
